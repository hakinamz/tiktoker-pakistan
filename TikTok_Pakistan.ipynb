```python
#import necessary liabraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Pakistan Top 100 TikTokers .csv')
df
```

```python
df = df['UserId,Name,Videos,Fans,Views,Account'].str.split(",",expand = True) 
df.head(5)
```

```python
df[2] = df[2]+df[3]
df
```

```python
df = df.drop(3, axis=1)
df
```

```python
df[6] = df[6]+' '+df[7]
df
```

```python
df = df.drop(7, axis=1)
df
```

```python
df = df.rename(columns={0: 'UserID', 1: 'Name', 2: 'Videos', 4: 'Fans', 5: 'Views', 6: 'Account'})
df.head()
```

```python
UserID = df['UserID'].str.split("@",expand = True)
UserID = '@'+UserID[1]
df['UserID'] = UserID
```

```python
df.head()
```

```python
def to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        if len(x) > 1:
            return float(x.replace('B', '')) * 1000000000
        return 1000000000.0
    
df['Fans'] = df['Fans'].apply(to_float)
df['Views'] = df['Views'].apply(to_float)
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
Videos = df['Videos'].str.split('"',expand = True)
Videos = Videos[1].astype('int32')
df['Videos'] = Videos
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.columns.unique()
```

```python
plt.figure(figsize=(17, 7))
plt.xticks(rotation=60)
graph = sns.barplot(x='UserID', y='Views', data=df.query('Views > 1.0e+09')) #1B
```

```python
plt.figure(figsize=(17, 7))
plt.xticks(rotation=60)
graph = sns.barplot(x='UserID', y='Fans', data=df.query('Fans > 10000000.0')) # 10M
```

```python
plt.figure(figsize=(10, 10))
dfvideo = df.query('Videos > 3000')
videos = list(dfvideo['Videos'])
names = list(dfvideo['UserID'])
plt.pie(videos, labels=names, autopct='%1.1f%%', colors=['#c4ff75', '#baff9e', '#ff9c59'])
plt.show()
```

```python
plt.figure(figsize=(10, 10))
dfViews = df.sort_values(by='Views', ascending=False)
videos = list(dfViews['Views'].head(10))
names = list(dfViews['UserID'].head(10))
colors = ['#baff9e', '#ff9c59', '#85dae6', '#e6859c', '#e69585', '#c1e685', '#87e685', '#e685dc', '#edabcc', '#c8e8c5']
plt.pie(videos, labels=names, autopct='%1.1f%%', colors=colors)
plt.show()
```
